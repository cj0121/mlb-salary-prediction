```python
# scatterplot: salary vs. position
plt.scatter(salaryFinalDF['position'], salaryFinalDF['avg_salary_log'], c='lightblue', alpha=0.5)
plt.show()
```

```python
# box plot: salary vs. position

data_1 = salaryFinalDF[salaryFinalDF['position'] == 'C']['avg_salary_log']
data_2 = salaryFinalDF[salaryFinalDF['position'] == '1B']['avg_salary_log']
data_3 = salaryFinalDF[salaryFinalDF['position'] == '2B']['avg_salary_log']
data_4 = salaryFinalDF[salaryFinalDF['position'] == 'SS']['avg_salary_log']
data_5 = salaryFinalDF[salaryFinalDF['position'] == '3B']['avg_salary_log']
data_6 = salaryFinalDF[salaryFinalDF['position'] == 'OF']['avg_salary_log']
data_7 = salaryFinalDF[salaryFinalDF['position'] == 'DH']['avg_salary_log']
data = [data_1, data_2, data_3, data_4, data_5, data_6, data_7]

fig = plt.figure(figsize =(9, 4))
ax = fig.add_axes([0, 0, 1, 1])

ax.set_xlabel('Position')
ax.set_ylabel('Avg. Salary Log')

bp = ax.boxplot(data, labels=["C", "1B", "2B", "SS", "3B", "OF", "DH"])
plt.show()
```
