# How many years of stats should we consider?

```python
data_1 = pd.read_csv('data/data_1.csv')
data_2 = pd.read_csv('data/data_2.csv')
data_3 = pd.read_csv('data/data_3.csv')
data_4 = pd.read_csv('data/data_4.csv')
data_5 = pd.read_csv('data/data_5.csv')
```

```python
predictors_full = ['Age', '1B_pos', '2B_pos', '3B_pos', 'C_pos',
       'DH_pos', 'SS_pos', 'G', 'PA', 'AB', 'R', 'H', '1B', '2B', '3B',
       'HR', 'TB', 'RBI', 'BB', 'SO', 'HBP', 'IBB', 'SB', 'CS', 'GDP',
       'SH', 'SF', 'AVG', 'OBP', 'SLG', 'OPS', 'ISO', 'BABIP', 'wRAA',
       'wRC+', 'WPA', 'WAR', 'Year_FA']
r2_results = []
mse_resultsTrain = []
mse_resultsTest = []
data_list = [data_1, data_2, data_3, data_4, data_5]
for i in range(5):
    r2_list = []
    mse_trainList = []
    mse_testList = []
    
    data = data_list[i]    
    kf = KFold(n_splits=5, random_state=0, shuffle=True)
    num_split = kf.get_n_splits(data)
    for train_index, test_index in kf.split(data):
        train_df = data.iloc[train_index]
        test_df = data.iloc[test_index]        
        results = func.fit_ols(predictors_full, 'Salary', train_df)
        # get R-squared on training sets
        r2 = results.rsquared
        r2_list.append(r2)
        # compute training mse
        resid = results.resid
        mse_train = sum(resid ** 2)/len(train_df)
        mse_trainList.append(mse_train)        
        # make prediction on test data
        y_pred = func.predict_ols(predictors_full, 'Salary', results, test_df)
        y_test = test_df['Salary']
        # get mse on prediction
        mse_test = sme.mse(y_pred, y_test, axis=0)
        mse_testList.append(mse_test)
    r2_results.append(round(sum(r2_list)/num_split, 3))
    mse_resultsTrain.append(round(sum(mse_trainList)/num_split, 0))
    mse_resultsTest.append(round(sum(mse_testList)/num_split, 0))
```

```python
x = [1,2,3,4,5]
fig, ax = plt.subplots(2)
fig.set_size_inches(8,5)
fig.tight_layout(pad=1.5)

sns.lineplot(x=x, y=mse_resultsTrain, ax=ax[0], legend='full', color='g')
sns.lineplot(x=x, y=mse_resultsTest, ax=ax[0], legend='full')
sns.lineplot(x=x, y=r2_results, ax=ax[1], color='y')
ax[0].set(ylabel="MSE", title='Mean Squared Errors')
ax[0].legend(['Train', 'Test'])
ax[1].set(xlabel='Number of Years Aggregated', ylabel="R-Squared Training", title='R-Squared of Fitted Model')
ax[0].axvline(x=3, ymin=0, ymax=1, color='r', dashes=[3])
ax[1].axvline(x=3, ymin=0, ymax=1, color='r', dashes=[3])

plt.setp(ax, xticks=[1,2,3,4,5], xticklabels=[1,2,3,4,5])
plt.show()
print(f"R-Squared(5-fold cross validated):")
for i in range(len(r2_results)):
    print(f"{i+1}: {r2_results[i]}")
```

## Examine Residual Plot of Fitted Model

```python
predictors_long = ['WAR', 'RBI', 'wRAA', 'Year_FA', 'Age', '3B', 'BB', '1B_pos', '2B_pos', '3B_pos', 'C_pos',
       'DH_pos', 'SS_pos']

results = func.fit_ols(predictors_long, 'Salary_sqrt', train_df)

fig, ax = plt.subplots()
fig.set_size_inches(6, 4)
sns.residplot(x=results.fittedvalues, y=results.resid, ax=ax,
                    lowess=True,
                    scatter_kws={'alpha': 0.5},
                    line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})
ax.set(ylabel='Residuals', xlabel='Fitted Values')
plt.show()
```

#### Examination:
1. The residuals across fitted values appear to be fairly symetrical, therefore, suggesting Linearity
2. The residuals increase as y increases, suggesting heteroscedasticity (unequal variances)
3. There appear to be a few outliers

## Assumption 1: Linearity

### x / y scatter plot

```python
fig, ax = plt.subplots(2, 4)
fig.set_size_inches(13, 6)
fig.tight_layout(pad=2)

count = 0
for a in range(2):
    for b in range(4):
        try:
            sns.scatterplot(x=train_df[predictors_short[count]], y=train_df['Salary_sqrt'], 
                            ax=ax[a, b], alpha=0.6)
            ax[a, b].set(ylabel='Salary_sqrt')
            count += 1
        except:
            continue
```

### Harvey Collier Test for linearity

H<sub>0</sub>: Regression is correctly modeled as linear.

H<sub>A</sub>: Non-linear relationship is presented between the response and predictors.

See [documentation](https://www.statsmodels.org/dev/generated/statsmodels.stats.diagnostic.linear_harvey_collier.html) for details.

Harvey, A. C., &amp; Collier, P. (1977). Testing for functional misspecification in regression analysis. Journal of Econometrics, 6(1), 103–119. https://doi.org/10.1016/0304-4076(77)90057-4 

```python
print ("          {:<9} {:<15}".format('t-stat','p-value'))
for i in range(len(predictors_short)):
    results1 = func.fit_ols(predictors_short[i], 'Salary_sqrt', train_df)
    hc = smd.linear_harvey_collier(results1, order_by=None, skip=3)
    print("{:<10}{:<10}{:<10}".format(predictors_short[i], round(hc[0],3),round(hc[1],3)))
```

# Assumption 2: Homoscedasticity

```python
fig, ax = plt.subplots(2, 4)
fig.set_size_inches(13, 6)
fig.tight_layout(pad=2)

count = 0
for a in range(2):
    for b in range(4):
        try:
            results2 = func.fit_ols(predictors_short[count], response, train_df)
            sns.scatterplot(x=train_df[predictors_short[count]], y=results2.resid, 
                            ax=ax[a, b], alpha=0.6)
            ax[a, b].set(ylabel='Residuals', xlabel=predictors_short[count])
            count += 1
        except:
            continue
```

## Breush-Pagan test:
H<sub>0</sub>: Homoscedasticity is present (the residuals are distributed with equal variance)

H<sub>A</sub>: Heteroscedasticity is present (the residuals are not distributed with equal variance)


See [documentation](https://www.statsmodels.org/stable/generated/statsmodels.stats.diagnostic.het_breuschpagan.html) from details.

```python
name = ["Lagrange multiplier statistic", "p-value", "f-value", "f p-value"]
test = smd.het_breuschpagan(results2.resid, results.model.exog, robust=True)
lzip(name, test)
```

# Assumption 3: Normally Distributed Errors

```python
fig, ax = plt.subplots(2, 4)
fig.set_size_inches(13, 6)
fig.tight_layout(pad=3)

count = 0
for a in range(2):
    for b in range(4):
        try:
            results3 = func.fit_ols(predictors_short[count], response, train_df)
            sns.histplot(data=results3.resid, ax=ax[a,b], bins=12, kde=True)
            ax[a, b].set(ylabel='Count', xlabel='Residuals', title=predictors_short[count])
            count += 1
        except:
            continue
```

### Anderson-Darling Normality Test

H<sub>0</sub>: The data follow normal distribution.

H<sub>A</sub>: The data do not follow normal distribution.

Penn State University. (n.d.). 7.5 - tests for error normality: Stat 501. Statistics Online Courses. Retrieved February 23, 2022, from https://online.stat.psu.edu/stat501/lesson/7/7.5 

National Institute of Standards and Technology. (n.d.). 1.3.5.14. Anderson-Darling Test. Engineering Statistics Handbook. Retrieved February 19, 2022, from https://www.itl.nist.gov/div898/handbook/eda/section3/eda35e.htm 

```python
print ("          {:<10}{:<10}".format('a-stat','p-value'))
for i in range(len(predictors_short)):
    results3 = func.fit_ols(predictors_short[i], response, train_df)
    ad = smd.normal_ad(results3.resid, axis=0)
    print("{:<10}{:<10}{:<10}".format(predictors_short[i], round(ad[0], 3), (round(ad[1], 3))))
#     print(f"{predictors_short[i]}:{round(ad[0], 3)}======{round(ad[1], 3)}")
```

```python
mse_list = []
rmse_list = []
rmspe_list = []
meanabs_list = []
medianabs_list = []
rate_list_low = []
rate_list_high = []
for train_index, test_index in kf.split(data_df):
    X_train = data_df.iloc[train_index][predictors_long]
    X_train = sm.add_constant(X_train)
    y_train = data_df.iloc[train_index][['Salary_sqrt']]
    model = sm.OLS(y_train, X_train)
    results_eval = model.fit()
    
    X_test = data_df.iloc[test_index][predictors_long]
    X_test = sm.add_constant(X_test)
    y_pred = results_eval.predict(X_test)
    y_test = data_df.iloc[test_index]['Salary_sqrt']
    
    y_pred = y_pred ** 2
    y_test = y_test ** 2
    
#     y_pred = np.exp(y_pred)
#     y_test = np.exp(y_test)
    
    
    diff = abs(y_test - y_pred) / y_test
    diff_df = diff.to_frame()
    
#     idx_low = y_test[y_test < y_test.median()].index
#     idx_high = y_test[y_test >= y_test.median()].index
    
    rate = len(diff_df[diff_df[0] <= 0.10])/len(y_test)
    
    
    mse = sme.mse(y_pred, y_test, axis=0)
    rmse = sme.rmse(y_pred, y_test, axis=0)
    rmspe = sme.rmspe(y_pred, y_test, axis=0)
    meanabs = sme.meanabs(y_pred, y_test, axis=0)
    medianabs = sme.medianabs(y_pred, y_test, axis=0)
    
    mse_list.append(mse)
    rmse_list.append(rmse)
    rmspe_list.append(rmspe)
    meanabs_list.append(meanabs)
    medianabs_list.append(medianabs)

    
avg_mse = round(sum(mse_list)/len(mse_list), 3)
avg_rmse = round(sum(rmse_list)/len(rmse_list), 3)
avg_rmspe = round(sum(rmspe_list)/len(rmspe_list), 3)
avg_meanabs = round(sum(meanabs_list)/len(meanabs_list), 3)
avg_medianabs = round(sum(medianabs_list)/len(medianabs_list), 3)


print(f"MSE: {avg_mse}; RMSE: {avg_rmse}; RMSPE: {avg_rmspe}; MEANABS: {avg_meanabs}; MEDIANABS: {avg_medianabs}")
```
