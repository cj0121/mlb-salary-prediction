# Assessing Model Assumptions


## Fit Model

```python
predictors_long = ['WAR', 'Age', 'Year_FA', '2B', 'wRAA', 'BB', 'HBP', 'RBI', 'SH', '1B_pos', '2B_pos', '3B_pos', 'C_pos',
       'DH_pos', 'SS_pos']

X = train_df[predictors_long]
y = train_df[response]
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

### Create Residual Plot of Fitted Model

```python
fig, ax = plt.subplots()
sns.scatterplot(x=results.fittedvalues, y=results.resid, ax=ax, alpha=0.6)
ax.set(ylabel='Residuals', xlabel='Fitted Values')
plt.show()
```

#### Examination:
1. The residuals across fitted values appear to be fairly symetrical, therefore, suggesting Linearity
2. The residuals increase as y increases, suggesting heteroscedasticity (unequal variances)
3. There appear to be a few outliers

## Assumption 1: Linearity

### x / y scatter plot

```python
fig, ax = plt.subplots(3, 3)
fig.set_size_inches(11, 10)
fig.tight_layout(pad=2)

count = 0
for a in range(3):
    for b in range(3):
        sns.scatterplot(x=train_df[predictors_short[count]], y=train_df['Salary_sqrt'], 
                        ax=ax[a, b], alpha=0.6)
        ax[a, b].set(ylabel='Salary_sqrt')
        count += 1
```

### Harvey Collier Test for linearity

```python
print ("          {:<9} {:<15}".format('t-stat','p-value'))
for i in range(len(predictors_short)):
    results1 = func.fit_ols(predictors_short[i], response, train_df)
    hc = smd.linear_harvey_collier(results1, order_by=None, skip=3)
    print("{:<10}{:<10}{:<10}".format(predictors_short[i], round(hc[0],3),round(hc[1],3)))
```

# Assumption 2: Homoscedasticity

```python
fig, ax = plt.subplots(3, 3)
fig.set_size_inches(9, 8)
fig.tight_layout(pad=2)

count = 0
for a in range(3):
    for b in range(3):
        results2 = func.fit_ols(predictors_short[count], response, train_df)
        sns.scatterplot(x=train_df[predictors_short[count]], y=results2.resid, 
                        ax=ax[a, b], alpha=0.6)
        ax[a, b].set(ylabel='Residuals', xlabel=predictors_short[count])
        count += 1
```

## Breush-Pagan test:

```python
name = ["Lagrange multiplier statistic", "p-value", "f-value", "f p-value"]
test = smd.het_breuschpagan(results2.resid, results.model.exog, robust=False)
lzip(name, test)
```

### SQRT

```python
X_test = test_df[predictors_long]
X_test = sm.add_constant(X_test)
y_pred = results.predict(X_test)
y_pred_sqrt = y_pred ** 2
y_true_sqrt = test_df['Salary_sqrt'] ** 2


mse_sqrt = sm.tools.eval_measures.mse(y_pred_sqrt, y_true_sqrt, axis=0)
rmse_sqrt = sm.tools.eval_measures.rmse(y_pred_sqrt, y_true_sqrt, axis=0)
rmspe_sqrt = sm.tools.eval_measures.rmspe(y_pred_sqrt, y_true_sqrt, axis=0)
meanabs_sqrt = sm.tools.eval_measures.meanabs(y_pred_sqrt, y_true_sqrt, axis=0)

print(f"MSE: {mse_sqrt}; RMSE: {rmse_sqrt}; RMSPE: {rmspe_sqrt}; MEANABS: {meanabs_sqrt}")
```

## 5 Fold CV

```python
kf = KFold(n_splits=2, random_state=0, shuffle=True)
kf.get_n_splits(data_df)
```

```python
mse_list = []
rmse_list = []
rmspe_list = []
meanabs_list = []
for train_index, test_index in kf.split(data_df):
    X_train = data_df.iloc[train_index][predictors_long]
    X_train = sm.add_constant(X_train)
    y_train = data_df.iloc[train_index][['Salary_sqrt']]
    model = sm.OLS(y_train, X_train)
    results = model.fit()
    
    X_test = data_df.iloc[test_index][predictors_long]
    X_test = sm.add_constant(X_test)
    y_pred = results.predict(X_test)
    y_test = data_df.iloc[test_index]['Salary_sqrt']
    
    mse = sme.mse(y_pred, y_test, axis=0)
    rmse = sme.rmse(y_pred, y_test, axis=0)
    rmspe = sme.rmspe(y_pred, y_test, axis=0)
    meanabs = sme.meanabs(y_pred, y_test, axis=0)
    
    mse_list.append(mse)
    rmse_list.append(rmse)
    rmspe_list.append(rmspe)
    meanabs_list.append(meanabs)
```

```python
avg_mse = sum(mse_list)/len(mse_list)
avg_rmse = sum(rmse_list)/len(rmse_list)
avg_rmspe = sum(rmspe_list)/len(rmspe_list)
avg_meanabs = sum(meanabs_list)/len(meanabs_list)
```

```python
print(f"MSE: {avg_mse}; RMSE: {avg_rmse}; RMSPE: {avg_rmspe}; MEANABS: {avg_meanabs}")
```
